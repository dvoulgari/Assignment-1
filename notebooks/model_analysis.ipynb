# Feature Selection

In this task, different feature selection method has been chosen accordingly to the models' needs.

## SVR

```python
model = SVR()
model.fit(X_train, y_train)
results = permutation_importance(model, X_train, y_train, scoring="neg_mean_squared_error", n_repeats=10, random_state=42)
importance = results.importances_mean
```

```python
# Plot feature importance
import matplotlib.pyplot as plt
from matplotlib import pyplot
# Create a bar chart of feature importance
pyplot.figure(figsize=(10, 6))
pyplot.title("Feature Importance - SVR")
pyplot.xlabel("Features")
pyplot.ylabel("Importance Score")
pyplot.bar([x for x in range(len(importance))], importance)
pyplot.show()
```

```python
# Sort features by importance
sorted_indices = importance.argsort()[::-1]
sorted_importance = importance[sorted_indices]

# Calculate cumulative importance
cumulative_importance = sorted_importance.cumsum() / sorted_importance.sum()

# Find the smallest number of features that explain 95% of the importance
selected_features_SVR = sorted_indices[cumulative_importance <= 0.95]
print("Selected features (indices):", selected_features_SVR)
optimal_num_features = (cumulative_importance <= 0.95).sum()
print(f"Optimal number of features: {optimal_num_features}")
np.save('selected_features_indices_SVR.npy', selected_features_SVR)

# Plot cumulative importance
plt.figure(figsize=(10, 6))
plt.title("Cumulative Feature Importance - SVR")
plt.xlabel("Number of Features")
plt.ylabel("Cumulative Importance Score")
plt.plot(range(1, len(cumulative_importance) + 1), cumulative_importance, marker='o')
plt.axhline(y=0.95, color='r', linestyle='--')
plt.axvline(x=optimal_num_features, color='g', linestyle='--')
plt.text(optimal_num_features + 1, 0.96, f"Optimal: {optimal_num_features}", color='g')
plt.show()
```

```python
# Train the model with selected features
SVR_model = SVR()
SVR_model.fit(
    X_train.iloc[:, selected_features_SVR], y_train)

# Evaluate the final model on training data
y_pred_train = SVR_model.predict(X_train.iloc[:, selected_features_SVR])

# Evaluate the final model on test data
y_pred_test = SVR_model.predict(X_test.iloc[:, selected_features_SVR])  

# Evaluate the final model on test data
y_pred_eval = SVR_model.predict(X_eval.iloc[:, selected_features_SVR]) 

# Calculate train metrics
rmse_train_SVR = np.sqrt(mean_squared_error(y_train, y_pred_train))
mae_train_SVR = mean_absolute_error(y_train, y_pred_train)
r2_train_SVR = r2_score(y_train, y_pred_train)

# Calculate test metrics
rmse_test_SVR = np.sqrt(mean_squared_error(y_test, y_pred_test))
mae_test_SVR = mean_absolute_error(y_test, y_pred_test)
r2_test_SVR = r2_score(y_test, y_pred_test)

# Calculate validation metrics
rmse_eval_SVR = np.sqrt(mean_squared_error(y_eval, y_pred_eval))
mae_eval_SVR = mean_absolute_error(y_eval, y_pred_eval)
r2_eval_SVR = r2_score(y_eval, y_pred_eval)

# Print train and test metrics
print(f"Final Model Performance (Train) - RMSE: {rmse_train_SVR}, MAE: {mae_train_SVR}, R2: {r2_train_SVR}")
print(f"Final Model Performance (Test) - RMSE: {rmse_test_SVR}, MAE: {mae_test_SVR}, R2: {r2_test_SVR}")
print(f"Final Model Performance (Validation) - RMSE: {rmse_eval_SVR}, MAE: {mae_eval_SVR}, R2: {r2_eval_SVR}")

# Save the final model
import joblib
joblib.dump(SVR_model, '/home/desp1/MLCB/models/SVR_FS.pkl')
```

# Hyperparameter tuning

```python
from sklearn.model_selection import GridSearchCV, KFold
from sklearn.metrics import make_scorer, mean_squared_error, mean_absolute_error, r2_score
import joblib
import numpy as np

# Load your pre-trained models
svr_model = joblib.load('SVR_FS.pkl')
eln_model = joblib.load('ElasticNet_FS.pkl')
br_model = joblib.load('BayesianRidge_FS.pkl')

# Define parameter grids for tuning
param_grids = {
    'SVR': {
        'C': [0.01, 0.1, 1, 10, 100], # wide range for better regularization
        'epsilon': [0.001, 0.01, 0.1, 0.5, 1.0], # sensitivity to outliers
        'kernel': ['linear', 'rbf'],
        'gamma': ['scale', 'auto', 0.1, 1.0] # used by RBF (non-linear) kernel
    },
    'ElasticNet': {
        'alpha': [0.001, 0.01, 0.1, 1],
        'l1_ratio': np.linspace(0, 0.9, 9), # more granularity
        'max_iter': [5000, 10000], # to avoid errors "You might want to increase the number of iterations", ensure convergance 
        'tol': [1e-4, 1e-5], # tighter tolerance to avoid errors
    },
    'BayesianRidge': {
       'alpha_1': [1e-7, 1e-5, 1e-4], # small values - more  precision
        'lambda_1': [1e-7, 1e-5, 1e-4], # wider range for noise
        'alpha_2': [1e-7, 1e-6, 1e-5],              
        'lambda_2': [1e-7, 1e-6, 1e-5],             
        'max_iter': [300, 500, 1000] # more iterations for convergence
    }
}


def tune_model(model, model_name, X_train, y_train, feature_indices, param_grid, cv=5):
    """Tune hyperparameters using feature indices"""
    # Convert to list if it's a set or numpy array
    if isinstance(feature_indices, (set, np.ndarray)):
        feature_indices = list(feature_indices)
    
    # Select features
    if hasattr(X_train, 'iloc'): 
        X_selected = X_train.iloc[:, feature_indices]
    else: 
        X_selected = X_train[:, feature_indices]
    
    # RMSE scorer for minimization
    from math import sqrt
    rmse_scorer = make_scorer(
        lambda y_true, y_pred: sqrt(mean_squared_error(y_true, y_pred)),
        greater_is_better=False
    )
    
    grid = GridSearchCV(
        estimator=model,
        param_grid=param_grid,
        cv=KFold(n_splits=cv, shuffle=True, random_state=42),
        scoring=rmse_scorer,
        n_jobs=-1,
        verbose = 1,
        pre_dispatch='2*n_jobs',  # prevent memory explosion
    )
    grid.fit(X_selected, y_train)
    
    print(f"\nBest parameters for {model_name}: {grid.best_params_}")
    print(f"Best RMSE: {-grid.best_score_:.4f}")
    
    return grid.best_estimator_, feature_indices

# Get your feature indices (replace with your actual variables)
selected_features_SVR = np.load("selected_features_indices_SVR.npy")
selected_features_ELN = np.load("selected_features_indices_ELN.npy")
selected_features_BR = np.load("selected_features_indices_BR.npy")

# Tune models
print("Tuning SVR...")
best_svr, svr_feats = tune_model(svr_model, 'SVR', X_train, y_train, 
                               selected_features_SVR, param_grids['SVR'])

print("\nTuning ElasticNet...")
best_eln, eln_feats = tune_model(eln_model, 'ElasticNet', X_train, y_train,
                               selected_features_ELN, param_grids['ElasticNet'])

print("\nTuning BayesianRidge...")
best_br, br_feats = tune_model(br_model, 'BayesianRidge', X_train, y_train,
                             selected_features_BR, param_grids['BayesianRidge'])
```

```python
def evaluate_all_models(models_dict, feature_indices_dict, X_sets, y_sets):
    """
    Evaluate models on all datasets (train, test, validation) and return metrics DataFrame
    
    Args:
        X_sets: {'Train': X_train, 'Test': X_test, 'Validation': X_val}
        y_sets: {'Train': y_train, 'Test': y_test, 'Validation': y_val}
    
    Returns:
        metrics_df: Pandas DataFrame with all metrics
    """
    results = []
    
    for model_name, (model, feat_indices) in models_dict.items():
        num_features = len(feat_indices)
        
        for stage, X in X_sets.items():
            y = y_sets[stage]
            
            # Select features
            if isinstance(X, pd.DataFrame):
                X_selected = X.iloc[:, feat_indices]
            else:
                X_selected = X[:, feat_indices]
            
            # Predict and calculate metrics
            y_pred = model.predict(X_selected)
            metrics = {
                'Model': model_name,
                'Stage': stage,
                'RMSE': np.sqrt(mean_squared_error(y, y_pred)),
                'MAE': mean_absolute_error(y, y_pred),
                'R2': r2_score(y, y_pred),
                'Num Features': num_features
            }
            results.append(metrics)
    
    return pd.DataFrame(results)

# Define your datasets
X_sets = {
    'Train': X_train,
    'Test': X_test,
    'Validation': X_eval  # Make sure you have validation set
}

y_sets = {
    'Train': y_train,
    'Test': y_test,
    'Validation': y_eval
}

# Create models dictionary with their feature indices
models_dict = {
    'SVR': (best_svr, svr_feats),
    'ElasticNet': (best_eln, eln_feats),
    'BayesianRidge': (best_br, br_feats)
}

# Get comprehensive metrics
metrics_df = evaluate_all_models(models_dict, 
                               {'SVR': svr_feats, 
                                'ElasticNet': eln_feats, 
                                'BayesianRidge': br_feats},
                               X_sets, y_sets)

# Print formatted table
from tabulate import tabulate
print(tabulate(metrics_df, headers='keys', tablefmt='psql', showindex=False))
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import RepeatedKFold
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import joblib

# Load all three models
models = {
    'ElasticNet': joblib.load('models/ElasticNet_final_model.pkl'),
    'SVR': joblib.load('models/SVR_final_model.pkl'),
    'BayesianRidge': joblib.load('models/BayesianRidge_final_model.pkl')
}

# Load evaluation data AND ensure same columns as training
train_columns = joblib.load('models/train_columns.pkl')  
eval_data = pd.read_csv('data/evaluation_final_data.csv')

# Reorder evaluation columns to match training
X_eval = eval_data[train_columns]
y_eval = eval_data['BMI']

# Evaluation function
def evaluate_model(model, X, y):
    try:
        y_pred = model.predict(X)
        return {
            'RMSE': np.sqrt(mean_squared_error(y, y_pred)),
            'MAE': mean_absolute_error(y, y_pred),
            'R2': r2_score(y, y_pred)
        }
    except ValueError as e:
        print(f"Prediction failed: {str(e)}")
        return None

# Evaluate each model
results = {}
for name, model in models.items():
    # Simple evaluation
    simple_results = evaluate_model(model, X_eval, y_eval)
    if simple_results is None:
        continue
        
    # Robust evaluation
    rkf = RepeatedKFold(n_splits=5, n_repeats=3, random_state=42)  # Fewer repeats for speed
    metrics = {'RMSE': [], 'MAE': [], 'R2': []}
    
    for train_idx, test_idx in rkf.split(X_eval):
        X_train, X_test = X_eval.iloc[train_idx], X_eval.iloc[test_idx]
        y_train, y_test = y_eval.iloc[train_idx], y_eval.iloc[test_idx]
        
        try:
            model.fit(X_train, y_train)
            fold_results = evaluate_model(model, X_test, y_test)
            if fold_results:
                for metric in metrics:
                    metrics[metric].append(fold_results[metric])
        except Exception as e:
            print(f"Fold failed for {name}: {str(e)}")
            continue
    
    # Store results
    results[name] = {
        'Simple': simple_results,
        'Robust': {
            'mean': {k: np.mean(v) for k,v in metrics.items()},
            'ci': {k: (np.percentile(v, 2.5), np.percentile(v, 97.5)) for k,v in metrics.items()}
        }
    }

# Print results
print("Model Evaluation Results:")
for name, res in results.items():
    print(f"\n{name}:")
    print("  Simple Evaluation:")
    print(f"    RMSE: {res['Simple']['RMSE']:.2f}")
    print(f"    MAE: {res['Simple']['MAE']:.2f}")
    print(f"    R2: {res['Simple']['R2']:.2f}")
    
    print("\n  Robust Evaluation (mean ± 95% CI):")
    for metric in ['RMSE', 'MAE', 'R2']:
        ci = res['Robust']['ci'][metric]
        print(f"    {metric}: {res['Robust']['mean'][metric]:.2f} ({ci[0]:.2f}-{ci[1]:.2f})")
```
